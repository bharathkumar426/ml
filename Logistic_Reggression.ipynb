```python
q = df['Pregnancies'].quantile(0.98)                               # 2% of data was removed
data_clean = df[df['Pregnancies'] < q]
```

```python
q = data_clean[ 'BMI' ].quantile(0.95)                             # 5% of data was removed
data_clean = data_clean[data_clean[ 'BMI' ]<q]
```

```python
q = data_clean['Glucose'].quantile(0.97)                           # 3% of data was removed
data_clean = data_clean[data_clean['Glucose']<q]
```

```python
q = data_clean['BloodPressure'].quantile(0.98)                     # 2% of data was removed
data_clean = data_clean[data_clean['BloodPressure']<q]
```

```python
q=data_clean['SkinThickness'].quantile(0.98)                       # 2% of data was removed
data_clean=data_clean[data_clean['SkinThickness']<q]
```

```python
q=data_clean['Insulin'].quantile(0.99)                             # 1% of data was removed
data_clean=data_clean[data_clean['Insulin']<q]
```

```python
q=data_clean['DiabetesPedigreeFunction'].quantile(0.90)            # 10% of data was removed
data_clean=data_clean[data_clean['DiabetesPedigreeFunction']<q]
```

```python
q=data_clean['Age'].quantile(0.99)                                  # 1% of data was removed
data_clean=data_clean[data_clean['Age']<q]
```

```python
df.shape         # Before quantile
```

```python
plt.figure(figsize=(20,10))
sns.boxplot(data_clean)
plt.show()
```

```python
X = data_clean.drop('Outcome',axis=1)
y = data_clean['Outcome']
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
log_reg=LogisticRegression()
```

```python
log_reg.fit(X_train,y_train)
```

```python
print(log_reg.coef_)
```

```python
print(log_reg.intercept_)
```

```python
log_pred=log_reg.predict(X_test)
```

```python
log_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,log_pred)
```

```python
log_reg.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,log_pred)
```
