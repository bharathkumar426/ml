```python
x=df.drop("PM 2.5",axis=1)
y=df["PM 2.5"]
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler.fit_transform(df)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42)
```

```python
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

```python
x_train
```

```python
y_train
```

```python
x_train_scaled
```

```python
x_test_scaled
```

```python
import scipy.cluster.hierarchy as sci
plt.figure(figsize=(20,7))
sci.dendrogram(sci.linkage(x_train_scaled,method='ward'))
plt.title("Dendograms")
```

```python
from sklearn.cluster import AgglomerativeClustering
agg_clustering = AgglomerativeClustering(n_clusters=2)
```

```python
agg_clustering.fit(x_train,y_train)
```

```python
agg_clustering.labels_
```

```python
agg_clustering.fit_predict(x_test)
```

```python
agg_pred=agg_clustering.fit_predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,agg_pred)
```

```python
from sklearn.metrics import silhouette_score
silhouette_avg = silhouette_score(x_test_scaled, agg_pred)
```

```python
print("silhouette score:",silhouette_avg*100)
```
